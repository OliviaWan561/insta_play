## Let's play!!!

### Login to instagram and get data
There is an ipython snippet where I do log in to my instagram account typing password and log in directly in code (_ha, what a lol_)


```python
from InstagramAPI import InstagramAPI
import numpy as np
import time


username = ''
pwd = ''


API = InstagramAPI(username,pwd)
API.login()
time.sleep(2)
```

Sometimes it's throwing me an error 429, but after that - "Login success!" So, let's get some data about our profile!

```python
API.getProfileData() # save data to LastJson attribute of API object
API.LastJson.keys() # LastJson - dict with all data from last function
```

> **When you invoke getters functions all received data collected in Last Json, which is a dict type instance.**

```python
print API.LastJson['user'].keys() # look what we have here
my_id = API.LastJson['user']['pk'] # profile id is 'pk'
```

First of all, we must know how much post's I have published:

```python
API.getUsernameInfo(my_id)
n_media = API.LastJson['user']['media_count']
```

Function **API.getUsernameInfo(user_id)** 
gives us more _specified_ information abut user, like *full_name* , *address_street*, *follower_count*, *edia_count* and others.

Let's get all media id's. This can be done step by step invoking **getUserFeed(usernameId, maxid)** function. Collected data - dict with information of post number, list of post's info (_items_), more_avaliable property and few other things.
For one step you can get data **only of 18** items because of instagram limits, maybe.
If there are no more post's - *more_available* equals False.

```python

media_ids = [] # container for all my post's id's
max_id = '' # some times instagram doesn't want to give you id's

for i in range(n_media/18+1): # getUserFeed() returns 18 posts or less
    API.getUserFeed(usernameId=my_id, maxid = max_id)
    media_ids += API.LastJson['items'] # if it's no post's further
    if API.LastJson['more_available']==False: # may be the end of feed
        print "no more avaliable"            # or just instagram limit's
        break
    max_id = API.LastJson['next_max_id'] # the first id of next iteration step
    print i, "   next media id = ", max_id, "  ", len(media_ids)
    time.sleep(3)
```

```python
print media_ids[0].keys() # info about media
```

Get ready to wait, because next part of code can take a lot of time! 

We will go throw all media and collect media likers:

```python
likers = []
m_id = 0
print "wait %.1f minutes" % (n_media*2/60.)
for i in range(len(media_ids)):
    m_id = media_ids[i]['id']
    API.getMediaLikers(m_id)
    likers += [API.LastJson]
    time.sleep(2)

print "done!"
```

Now we need to sort users and count all likes of unique user:

```python
users = []
for i in likers:
    users += map(lambda x: i['users'][x]['username'], range(len(i['users'])))

# users - all users that likes my media [username]*num_liked_posts(username)
users_set = set(users)
# users_set - unique users that likes my media
print "all users = ", len(users), "  uniqum users = ", len(users_set)
```

For me, there are 6428 likes from 1060 accounts.

Make a dict with information of likes per user


```python
l_dict = {}
for user in users_set:
    l_dict[user] = users.count(user)  # {username:number_of_liked_posts}
```

The plotting part:

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
import operator


all_pairs = sorted(l_dict.items(), key=operator.itemgetter(1))
n_users = 10 # top 10 users

pairs = all_pairs[-n_users:]

y = map(lambda y: pairs[y][1], range(len(pairs)))
x = map(lambda y: pairs[y][0], range(len(pairs)))

```

**all_pairs** - sorted list of tuples with structure \[(username, num_likes),...\]
We'll take only 10 top followers and make a plot:

```python
fig = plt.figure()
plt.xkcd()
plt.xticks(range(len(pairs)), x, rotation='vertical')
plt.ylim([80, 160])
plt.bar(range(len(pairs)), y)
plt.xlabel('USERS')
plt.ylabel('number of liked posts')
plt.show()
```

```python
plt.gcf().clear()

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
explode = map(lambda i: i/max(y)/10., y)

fig1, ax1 = plt.subplots()
ax1.pie(y, explode=explode, labels=x, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
